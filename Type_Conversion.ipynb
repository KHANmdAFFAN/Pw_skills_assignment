### Question 4

```python
list1 = [i**2 for i in range(20) if i%2==0]
print(f"List {list1}")
print(f"List to tuple --- > {tuple(list1)}")
```

### Question 5

```python
tuple1 = tuple(filter(lambda x:x%2==0,[i for i in range(20)]))
print(f"tuple ---> {tuple1}")
print(f"list to  tuple ----> {list(tuple1)}")
```

### Question 6

```python
def decimal_to_binary(x):
    d2b = bin(x)[2:]
    return d2b
x = int(input("Enter the number to change in binary "))
decimal_to_binary(x)
```
