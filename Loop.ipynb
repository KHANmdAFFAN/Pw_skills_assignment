```python
# Define the range of the multiplication table
start_range = 1
end_range = 10

# Outer loop for the multiplicand
for i in range(start_range, end_range + 1):
    # Inner loop for the multiplier
    for j in range(1, 11):
        result = i * j
        print(f"{i} x {j} = {result}")

    # Add a newline after each row in the table
    print()
```
