### Question 1

```python
def calculate(a,b,operator):
    if operator is  "+":
        sum1 = a + b
        print(sum1)
    elif operator is "-":
        diff = a - b
        print(diff)
    elif operator is "x":
        product = a * b
        print(product)
    elif operator is "/":
        if b!=0:
            que = a/b
            print(que)
        else:
            print("division by zero")
    return
            
        
a = int(input("Enter First number : "))
b = int(input("Enter second number : "))
operator = input("Enter which operation You want to choose :")    

calculate(a,b,operator)    
```

### Question 6

```python
num1 = int(input("Enter the first integer: "))
num2 = int(input("Enter the second integer: "))

bit_and = num1 & num2
bit_or = num1 | num2
bit_xor = num1 ^ num2
left_shift = num1 << num2
right_shift = num1 >> num2

print(f"Bitwise AND: {bit_and}")
print(f"Bitwise OR: {bit_or}")
print(f"Bitwise XOR: {bit_xor}")
print(f"Left Shift: {left_shift}")
print(f"right Shift: {right_shift}")
```
